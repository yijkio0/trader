```python
def bearish_reversal(data):
    global shooting_star,bearish_engulfing,advance_block,hanging_man,tweezer_top,evening_star
    if data["close"].iloc[-3]<data["close"].iloc[-2]<data["close"].iloc[-1]:  # Shooting Star
        if data["color"].iloc[-1]==0: 
            if data["upper_wick"].iloc[-1]>2*data["abs_body"].iloc[-1]:
                if data["lower_wick"].iloc[-1]<0.2*data["upper_wick"].iloc[-1]:
                    shooting_star+=1
                    return 1
    if data["color"].iloc[-2]==1:
        if data["color"].iloc[-1]==0:
            if data["abs_body"].iloc[-1]>data["abs_body"].iloc[-2]:
                    bearish_engulfing+=1
                    return 2
    if all(data["color"].iloc[-3:]==1):  # Advance Block
        if data["body"].iloc[-3]>data["body"].iloc[-2]:
            if data["body"].iloc[-2]>data["body"].iloc[-1]:
                advance_block+=1
                return 3
    if data["color"].iloc[-1]==0:  # Hanging Man
        if data["lower_wick"].iloc[-1]>2*data["abs_body"].iloc[-1]:
            if data["upper_wick"].iloc[-1]<0.2*data["lower_wick"].iloc[-1]:
                hanging_man+=1
                return 4
    if data["color"].iloc[-1]==0:  # Tweezer Top
        if data["color"].iloc[-2]==1:
            if data["abs_body"].iloc[-1]<data["abs_body"].iloc[-2]:
                if 0.99*data["high"].iloc[-1]<data["high"].iloc[-2]<1.01*data["high"].iloc[-1]:
                    tweezer_top+=1  
                    return 5
    if data["color"].iloc[-3]==1:  # Evening Star
        if data["color"].iloc[-1]==0:
            if data["upper_wick"].iloc[-3]+data["lower_wick"].iloc[-3]<data["abs_body"].iloc[-3]:
                if data["upper_wick"].iloc[-1]+data["lower_wick"].iloc[-1]<data["abs_body"].iloc[-1]:
                    if data["abs_body"].iloc[-2]<0.2*data["upper_wick"].iloc[-2]:
                        if data["abs_body"].iloc[-2]<0.2*data["lower_wick"].iloc[-2]:
                            evening_star+=1
                            return 6
    return 0
```

```python
def bullish_reversal(data):
    global hammer,inverted_hammer,bullish_engulfing,three_stars_in_the_south,tweezer_bottom,morning_star
    if data["color"].iloc[-1]==1:  # Hammer
        if data["lower_wick"].iloc[-1]>2*data["body"].iloc[-1]:
            if data["upper_wick"].iloc[-1]<0.2*data["lower_wick"].iloc[-1]:
                hammer+=1
                return 1
    if data["color"].iloc[-1]==1:  # Inverted Hammer
        if data["upper_wick"].iloc[-1]>2*data["body"].iloc[-1]:
            if data["lower_wick"].iloc[-1]<0.2*data["upper_wick"].iloc[-1]:
                inverted_hammer+=1
                return 2
    if data["color"].iloc[-2]==0:  # Bullish Engulfing
        if data["abs_body"].iloc[-2]>data["upper_wick"].iloc[-2]:
            if data["abs_body"].iloc[-2]>data["lower_wick"].iloc[-2]:
                if data["color"].iloc[-1]==1:
                    if data["body"].iloc[-1]>data["upper_wick"].iloc[-1]:
                        if data["body"].iloc[-1]>data["lower_wick"].iloc[-1]:
                            if data["body"].iloc[-1]>data["body"].iloc[-2]:
                                bullish_engulfing+=1
                                return 3
    if all(data["color"].iloc[-3:]==0):  # Three Stars in the South
        if data["abs_body"].iloc[-3]>data["abs_body"].iloc[-2]:
            if data["abs_body"].iloc[-2]>data["abs_body"].iloc[-1]:
                three_stars_in_the_south+=1
                return 4
    if data["color"].iloc[-1]==1:  # Tweezer Bottom
        if data["color"].iloc[-2]==0:
            if data["abs_body"].iloc[-1]>data["abs_body"].iloc[-2]:
                if 0.9*data["lower_wick"].iloc[-1]<data["lower_wick"].iloc[-2]<1.1*data["lower_wick"].iloc[-1]:
                    tweezer_bottom+=1
                    return 5
    if data["color"].iloc[-3]==0:  # Morning Star
        if data["color"].iloc[-1]==1:
            if data["upper_wick"].iloc[-3]+data["lower_wick"].iloc[-3]<data["abs_body"].iloc[-3]:
                if data["upper_wick"].iloc[-1]+data["lower_wick"].iloc[-1]<data["abs_body"].iloc[-1]:
                    if data["abs_body"].iloc[-2]<0.2*data["upper_wick"].iloc[-2]:
                        if data["abs_body"].iloc[-2]<0.2*data["lower_wick"].iloc[-2]:
                            morning_star+=1
                            return 6
    return 0
```

```python
def bearish_continuation(data):
    global bearish_three_line_strike,falling_three_methods,bearish_mat_hold
    if data["color"].iloc[-4]==0:  # Bearish Three Line Strike 
        if data["color"].iloc[-3]==0:   
            if data["color"].iloc[-2]==0:
                if data["color"].iloc[-1]==1:
                    if data["abs_body"].iloc[-1]>(data["abs_body"].iloc[-4]+data["abs_body"].iloc[-3]+data["abs_body"].iloc[-2]):
                        bearish_three_line_strike+=1
                        return 1
    if data["color"].iloc[-5]==0:  # Falling Three Methods 
        if data["color"].iloc[-4]==1:
            if data["color"].iloc[-3]==1:
                if data["color"].iloc[-2]==1:
                    if data["color"].iloc[-1]==0:
                        if data["lower_wick"].iloc[-5]+data["upper_wick"].iloc[-5]<0.5*data["abs_body"].iloc[-5]:
                            if data["close"].iloc[-2]-data["open"].iloc[-4]<data["open"].iloc[-5]-data["close"].iloc[-5]:
                                if data["close"].iloc[-1]<data["close"].iloc[-5]:
                                    falling_three_methods+=1
                                    return 2
    if data["color"].iloc[-5]==0:  # Bearish Mat Hold
        if data["color"].iloc[-4]==1:
            if data["color"].iloc[-3]==1:
                if data["color"].iloc[-2]==1:
                    if data["color"].iloc[-1]==0:
                        if data["lower_wick"].iloc[-5]+data["upper_wick"].iloc[-5]<0.5*data["abs_body"].iloc[-5]:
                            if data["open"].iloc[-4]<data["close"].iloc[-5]:
                                if data["high"].iloc[-2]-data["low"].iloc[-4]<data["high"].iloc[-5]-data["low"].iloc[-5]:
                                    if data["close"].iloc[-1]<data["low"].iloc[-5]:
                                        bearish_mat_hold+=1
                                        return 3
    return 0
```

```python
def bullish_continuation(data):
        global bullish_three_line_strike,rising_three_methods,bullish_mat_hold
        if data["color"].iloc[-4]==1:  # Bullish Three Line Strike 
            if data["color"].iloc[-3]==1:
                if data["color"].iloc[-2]==1:
                    if data["color"].iloc[-1]==0:
                        if data["abs_body"].iloc[-1]>(data["abs_body"].iloc[-4]+data["abs_body"].iloc[-3]+data["abs_body"].iloc[-2]):
                            bullish_three_line_strike+=1
                            return 1
        if data["color"].iloc[-5]==1:  # Rising Three Methods 
            if data["color"].iloc[-4]==0:
                if data["color"].iloc[-3]==0:
                    if data["color"].iloc[-2]==0:
                        if data["color"].iloc[-1]==1:
                            if data["lower_wick"].iloc[-5]+data["upper_wick"].iloc[-5]<0.5*data["abs_body"].iloc[-5]:
                                if data["open"].iloc[-4]-data["close"].iloc[-2]<data["close"].iloc[-5]-data["open"].iloc[-5]:
                                    if data["close"].iloc[-1]>data["close"].iloc[-5]:
                                        rising_three_methods+=1
                                        return 2 
        if data["color"].iloc[-5]==1:  # Bullish Mat Hold
            if data["color"].iloc[-4]==0:
                if data["color"].iloc[-3]==0:
                    if data["color"].iloc[-2]==0:
                        if data["color"].iloc[-1]==1:
                            if data["lower_wick"].iloc[-5]+data["upper_wick"].iloc[-5]<0.5*data["abs_body"].iloc[-5]:
                                if data["open"].iloc[-4]>data["close"].iloc[-5]:
                                    if data["high"].iloc[-4]-data["low"].iloc[-2]<data["high"].iloc[-5]-data["low"].iloc[-5]:
                                        if data["close"].iloc[-1]>data["high"].iloc[-5]:
                                            bullish_mat_hold+=1
                                            return 3
        return 0
```

```python
data=[]
api_key='e62c395ec9844095ada4b688d012a975'
technical='time_series'
ticker='BTC/USD'
timezone='Asia/Calcutta'
interval='30min'
for yr in range(2022,2026):
    for m in range(1,5): 
        last_day=calendar.monthrange(yr,3*m)[1]   
        start_date=f'{yr}-{(3*m-2):02d}-01T00:00:00'
        end_date=f'{yr}-{(3*m):02d}-{last_day}T23:30:00'
        api_url=f'https://api.twelvedata.com/{technical}?symbol={ticker}&order=asc&timezone={timezone}&start_date={start_date}&end_date={end_date}&interval={interval}&outputsize=5000&apikey={api_key}'
        while True:
            session=requests.Session()
            retries=Retry(total=5,backoff_factor=1,status_forcelist=[429,500,502,503,504])
            session.mount("https://",HTTPAdapter(max_retries=retries))
            fetch=session.get(api_url).json()
            if "values" in fetch:
                df=pd.DataFrame(fetch['values'])
                data.append(df)
                time.sleep(8)
                break
            elif "code" in fetch and fetch["code"]==429:
                print("Rate limit exceeded")
                time.sleep(15)
            else: 
                print("API Error:",fetch)
                fetch=None
                break
data=pd.concat(data,ignore_index=True)
data['color']=np.where(data['close']>data['open'],1,0)
data['open']=pd.to_numeric(data['open'])
data['close']=pd.to_numeric(data['close'])
data['abs_body']=abs(data['close']-data['open'])
data['body']=(data['close']-data['open'])
data['high']=pd.to_numeric(data['high'])
data['low']=pd.to_numeric(data['low'])
data['upper_wick']=np.where(data['color']==1,data['high']-data['close'],data['high']-data['open'])
data['lower_wick']=np.where(data['color']==1,data['open']-data['low'],data['close']-data['low'])
data=data.dropna()
data=data.reset_index(drop=True)
data
```
